# Modelo matemático del crecimiento y tratamiento de un tumor

$$\frac{dc}{dt} = -\lambda\ c\log \frac{c}{K} \; $$

$$\frac {dK}{dt} =\phi c  -\varphi k c^\frac {2}{3} $$

 _c_ representa al tamaño del tumor a través del tiempo 
 
 _k_ representa al capacidad de carga del tumor
 
 Parámetros:
 
$\phi = 5.85 $ 

$\lambda = 0.192 $ 

$ \varphi = 0.00873 $  

## Simulación

```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.integrate import odeint


```

```python
def crecimiento_tratamiento(x,t,phi,rho,lamda,XIval):
    x1 = x[0]
    x2 = x[1]
    if t<40:
        XI = 0
    else:
        XI = XIval
    return [-lamda*x1*np.log10(x1/x2)-XI*x1,
             phi*x1-rho*x2*(x1**(2/3))]
```

```python
## condiciones iniciales
phi = 5.85#parametro
rho = 0.00873#parametro
lamda=0.192#parametro
XI = 0.1 # condicion iniail ? Valor arbitrario
x0 = [200, 500]
t = np.linspace(0,61,10000)
```

```python
x = odeint(crecimiento_tratamiento, x0, t, args=(phi,rho,lamda,XI))
```

```python
x
```

```python
#Gráficas
plt.figure(figsize=(8,6))
plt.plot(t, x[:,0],lw=2,label='Evolución del tamaño del tumor')
plt.plot(t, x[:,1],lw=2,label='Capacidad de carga de las células')
plt.grid(True, which="both")
plt.legend(loc='best')
plt.title('Reacción al tratamiento')
plt.xlabel('Días')
plt.ylabel('Tamaño del tumor')

plt.show()
```

```python
def manip_medicamento(XI = 0):
    ## condiciones iniciales
    phi = 5.85#parametro
    rho = 0.00873#parametro
    lamda=0.192#parametro
    x0 = [200, 500]
    t = np.linspace(0,61,10000)
    x = odeint(crecimiento_tratamiento, x0, t, args=(phi,rho,lamda,XI)) 
    
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(t, x[:,0],lw=2,label='Evolución del tamaño del tumor')
    ax.plot(t, x[:,1],lw=2,label='Capacidad de carga de las células')
    plt.grid(True, which="both")
    plt.legend(loc='best')
    plt.title('Reacción al tratamiento')
    plt.xlabel('Días')
    plt.ylabel('Tamaño del tumor')

    fig.canvas.draw()

interact_manual(manip_medicamento, XI = (0, 1, .05));
```
